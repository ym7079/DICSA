## 1.데이터셋 불러오기 연습

다음 url에서 엑셀 파일을 불러오세요. (pandas excel)

<https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx>

  - 데이터셋 변수의 이름은 `df1`, `df2`로 저장하세요.
  - row가 아닌 column에 '매출액', '영업이익' 이 가도록 데이터를 transpose 하세요.
  - 이후 아래 링크를 참조하여 첫번째 row를 column name으로 설정하세요.

<https://stackoverflow.com/questions/31328861/python-pandas-replacing-header-with-top-row>

- 파일은 2개의 시트로 구성되어 있으며 각각 이름은 `008770`, `035250` 입니다.



````python
excel_url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx'
#데이터셋 변수의 이름은 df1, df2로 저장
df1 = pd.read_excel(excel_url, sheet_name = '008770 ').transpose()
df2 = pd.read_excel(excel_url, sheet_name='035250').transpose()

header_df1 = df1.iloc[0]
header_df2 = df2.iloc[0]

df1 = df1[1:]
df2 = df2[1:]

df1.columns = header_df1
df2.columns = header_df2

df1
````

## 2. 데이터 확인

- 데이터의 dimension을 확인하세요.
- 데이터의 결측치가 어디에 있는지 확인 후 0으로 대체하세요.
- 각 데이터 별로 **FCF**의 값을 barplot으로 시각화하세요.
  - 0보다 큰값, 작은값의 수의 총합을 `num1`, `num2`에 저장하세요


````python
#데이터 dimension 확인
print(df1.shape)
print(df2.shape)
````

````python
#데이터의 결측치가 어디에 있는지 확인 
pd.DataFrame(df1.isnull().sum(), columns=["결측치"])
````

````python
pd.DataFrame(df2.isnull().sum(), columns=["결측치"])
````

````python
#데이터의 결측치 0으로 대체
df1 = df1.fillna(0)
df2 = df2.fillna(0)

#0으로 대체되었는지 확인(df1)
df1
````

````python
#0으로 대체되었는지 확인(df2)
df2
````

````python
#각 데이터 별로 FCF의 값을 barplot으로 시각화
df1['FCF'].plot.bar();
````

````python
df2['FCF'].plot.bar();
````

````python
#0보다 큰값, 작은값의 수의 총합을 num1, num2에 저장
# 0보다 큰값
num1_df1 = df1.FCF[df1['FCF'] > 0].value_counts().sum()
num1_df2 = df2.FCF[df2['FCF'] > 0].value_counts().sum()
num1 = num1_df1 + num1_df2
print(num1)
````

````python
# 0보다 작은값
num2_df1 = df1.FCF[df1['FCF'] < 0].value_counts().sum()
num2_df2 = df2.FCF[df2['FCF'] < 0].value_counts().sum()
num2 = num2_df1 + num2_df2
print(num2)
````

````python
import seaborn as sns
 
pp = sns.load_dataset('penguins')
pp
````

````python
import matplotlib.pyplot as plt

#pp의 species와 sex변수에 대해서 cross-tabulation을 시행
pp_crosstab = pd.crosstab(index=pp['species'],columns=pp['sex'])
pp_crosstab
````
